```python
import pandas as pd
def RidgeReg3(X,y,lam):
    #X = np.array([[1,2,3],[4,5,6],[7,10,9],[6,2,3]])
    #y = np.array([[12,32,43]])
    b,c =X.shape
    #print(b)
    a=np.ones((1,c))
    X_1=np.vstack((X,a))
    X_tran = np.transpose(X_1)
    C = np.dot(X_1,X_tran)

    I = np.identity(b)
    z = np.zeros((b,1))
    I = np.hstack((I,z))
    z = np.transpose(np.zeros((b+1,1)))
    I_bar = np.vstack((I,z))
    #I_bar

    I_bar=I_bar*lam
    X_left = np.add(C,I_bar)
    #print(y.shape)
    y = y[:3001]
    X_right=np.dot(X_1,y)
    theta = np.linalg.lstsq(X_left,X_right)[0]
    #print(type(theta))
    theta = pd.DataFrame(list(theta))
    #print(theta.shape)
    theta_1 = pd.DataFrame(theta[0])
    print(theta_1.shape)
    theta_2 = theta_1.as_matrix()
    print(theta_2.shape)
    bias = theta_2[b][0]
    
    theta_3=np.delete(theta_2,-1)
    theta_3 = np.reshape(theta_3,(4999,1))
    print(theta_3.shape)
    return theta_3,bias
    #t_theta = np.dot(theta_3,X)+bias
f1,f2 = RidgeReg3(X_train,y_train[y_train.columns[1]],1000)
f1
```
